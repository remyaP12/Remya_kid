```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Example data for this case study
data = {
    'Age': [25, 40, 35, 50, 23, 45, 30, 60, 27, 39 ],
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Education': ['Diploma','Bachelor', 'Master', 'PhD', 'Diploma', 'Master', 'PhD', 'Bachelor', 'Master', 'PhD'],
    'Industry': ['Manufacturing', 'IT', 'Manufacturing', 'Healthcare', 'Manufacturing', 'IT', 'Healthcare', 'IT', 'Manufacturing', 'IT'],
    'Residence': ['Metro','Non-metro', 'Metro', 'Metro', 'Non-metro', 'Metro', 'Non-metro', 'Metro', 'Metro', 'Non-metro' ],
    'Income Class': ['Below 4L', '4-15L', '4-15L', 'Above 15L', 'Below 4L', '4-15L', 'Above 15L', '4-15L', 'Below 4L', '4-15L']
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)
df
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder


# Initialize LabelEncoder
le = LabelEncoder()

# Label Encoding the target variable 'Income Class'
df['Income Class'] = le.fit_transform(df['Income Class'])

# Check the classes that were encoded by LabelEncoder
print("Classes for 'Income Class':", le.classes_)

# Label Encoding the features (Gender, Education, Industry, Residence)
df['Gender'] = le.fit_transform(df['Gender'])
df['Education'] = le.fit_transform(df['Education'])
df['Industry'] = le.fit_transform(df['Industry'])
df['Residence'] = le.fit_transform(df['Residence'])

# Define the individual data (for prediction)
individual_data = {
    'Age': [35],
    'Gender': [1],       # Assuming 'Male' is encoded as 1
    'Education': [0],    # Assuming 'Diploma' is encoded as 0
    'Industry': [0],     # Assuming 'Manufacturing' is encoded as 0
    'Residence': [1]     # Assuming 'Metro' is encoded as 1
}
# Convert the dictionary to a DataFrame
individual_df = pd.DataFrame(individual_data)

# Prepare training data (X) and target variable (y)
x = df.drop(columns=['Income Class'])
y = df['Income Class']

# Splitting data into training and testing sets (80% training, 20% testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(x_train, y_train)

# Ensure the individual data has the same columns as the training data
individual_df = individual_df[[col for col in x_train.columns if col in individual_df.columns]]

# Predicting the income class for the individual
prediction = rf_classifier.predict(individual_df)

# Decoding the prediction back to the original income class labels
predicted_income_class = le.inverse_transform(prediction)

# Output the result
print(f"Predicted income class for the individual: {predicted_income_class[0]}")

```
